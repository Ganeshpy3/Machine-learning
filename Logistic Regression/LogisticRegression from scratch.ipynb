```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("heart.csv")
data.head()
```

```python
Y_val=data["target"]
```

```python
X_val=data.drop(["target"],axis=1)
```

```python
X_val
```

```python

from sklearn.model_selection import train_test_split
```

```python
def sigmoid(Y):

    return 1/(1+np.exp(-Y))
```

```python
def  cost_function(X,Y,learning_rate,iteration):
    n=X.shape[1]
    m=X.shape[0]
    W=np.zeros(n)
    B=0
    for i in range(iteration):
        Z=np.dot(W,X.T)+B
        a=sigmoid(Z)
        dw=(1/m)*np.dot(a-Y, X)
        db=(1/m)*np.sum(a - Y)
        W = W - learning_rate*dw.T
        B = B - learning_rate*db
        
        
    return W,B
    
    
    
```

```python
Xtrain,xtest,ytrain,ytest=train_test_split(X_val,Y_val,test_size=0.3)
len(ytrain)
print(Xtrain.shape)
print(ytrain.shape)
```

```python

m,b=cost_function(Xtrain,ytrain,0.01,100000)
```

```python
xp=np.dot(xtest,m)+b
```

```python
z=sigmoid(xp)
```

```python
for i in range(len(z)):
    if z[i]>=0.5:
        z[i]=1
    else:
        z[i]=0
```

```python
a=z==ytest
```

```python
c=0
for i in a:
    if i == False:
        c=c+1
```

```python
accuracy=((len(ytest)-c)/len(ytest))*100
```
