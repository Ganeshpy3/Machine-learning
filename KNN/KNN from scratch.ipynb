```python
Y=df["diagnosis"]
```

```python
X=df.drop(columns=["diagnosis"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
def distance(df,xtest):
    li=[]
    for i in df-xtest:
        li.append((sum(i**2))**0.5)
    return li
```

```python
def sort(val,k):
    newlist=val.copy()
    val.sort()
    inde=[]
    for i in val[:k]:
        inde.append(newlist.index(i))
    return inde
        
```

```python
def result(ytrain,inde):
    op=[]
    for i in inde:
        op.append(ytrain.values[i])
    c={}
    for i in op:
        if i in c:
            c[i]+=1
        else:
            c[i]=1
    return c
        
        
```

```python
def predicted_val(di):
    max=0
    for i in di:
        if di[i]>max:
            max=di[i]
            val=i
    return val
```

```python
def accuracy(score):
    scoreval=0
    for i in score:
        if i == True:
            scoreval +=1
    accuracy=(scoreval/len(score))*100
    print("Accuracy :{:.2f}% ".format(accuracy))
    
```

```python
def predict(Xtrain,Xtest,ytrain):
    j=0
    score=[]
    df=Xtrain.values
    for i in xtest.values:
        lis=distance(df,i)
        di=result(ytrain,sort(lis,5))
        print("Actual ->{}-----> Predicted ->{} ".format(ytest.values[j],predicted_val(di)))
        score.append(ytest.values[j]==predicted_val(di))
        j=j+1
    accuracy(score)
    
    
```

```python
predict(xtrain,xtest,ytrain)
```
