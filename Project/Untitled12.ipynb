```python
totalstop=[]
for i in df["Total_Stops"]:
    if i == "non-stop":
        totalstop.append(0)
    elif i == "1 stop":
        totalstop.append(1)
   
    elif i == "2 stops":
        totalstop.append(2)
    
    elif i == "3 stops":
        totalstop.append(3)
    elif i=="4 stops":
        totalstop.append(4)
        
    
```

```python
total_stop=pd.DataFrame(totalstop,columns=["stops"])
total_stop
```

```python
total_stop["stops"].value_counts()
```

```python
df["Total_Stops"].value_counts()
```

```python
df["Total_Stops"]=total_stop
```

```python
df
```

```python
minutes=np.zeros((df.shape[0],1))
for idx, i in enumerate(df["Duration"]):
    splitval=i.split("h")
    if len(splitval)==2:
        hr=int(splitval[0])*60
        mi=splitval[1].split("m")[0]
        if len(mi)>1:
            hr +=int(mi)
    else:
        hrs=splitval[0].split("m")[0]
        hr=int(hrs)
        print(idx)
    minutes[idx]=hr
    
```

```python
int(mi)
```

```python
hrs[0].split("m")[0]
```

```python
df["Minutes"]=pd.DataFrame(minutes)
```

```python
time=np.full((updatedprice.shape[0],1),"null              ")
for idx, i in enumerate(updatedprice["Dep_Time"]):
    val=i.split(":")[0]    
    if int(val)>=0 and int(val) <12:
        time[idx]="Morning"
    elif int(val)>=12 and int(val)<18:
        time[idx]="Afternoon"
    elif int(val)>=18 and int(val)<=23:
        time[idx]="Night"
     
```

```python
Time=pd.DataFrame(time)
```

```python
Time
```

```python
updatedprice["Time"]=Time
```
