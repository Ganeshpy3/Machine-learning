```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
sk=StratifiedKFold(5)
```

```python
for i,j in sk.split(X,Y):
    print("Testing data")
    countte=0
    
    lengthte=len(Y)
    for s in Y[j]:
        if s==1:
            countte+=1
    print("1",countte)
    print("0",lengthte-countte)

    print("Training data")
    counttr=0
    lengthtr=len(Y)
    for s in Y[i]:
        if s==1:
            counttr+= 1
    print("1",counttr)
    print("0",lengthtr-counttr)
    
    
```

```python
model1= LogisticRegression(solver='liblinear')
```

```python
result1=cross_val_score(model1,X,Y,cv=sk)
```

```python
print(result1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model2=KNeighborsClassifier(3)
```

```python
result2=cross_val_score(model2,X,Y,cv=sk)
print(result2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
```

```python
result3=cross_val_score(model3,X,Y,cv=sk)
print(result3)
```

```python
print("Logistic Regression model accuracy",np.mean(result1))
print("KNN model accuracy",np.mean(result2))
print("Decision tree  model accuracy",np.mean(result3))
```
