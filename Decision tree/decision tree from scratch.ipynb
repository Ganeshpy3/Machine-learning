```python
def classcount(data):
    target=data[:,-1]
    class_count={}
    for i in target:
        if i in class_count:
            class_count[i] +=1
        else:
            class_count[i]=1
    return class_count
```

```python
class question:
    def __init__(self,column,value):
        self.column=column
        self.value=value
        
    def match(self,example):
        val=example[self.column]
        if isnumeric(val):
            return val>=self.value
        else:
            return val == self.value
    def __repr__(self):
        condition = "=="
        if isnumeric(self.value):
            condition = ">="
        return "Is %s %s %s?" % (
            header[self.column], condition, str(self.value))

        
```

```python
a=question(3,0.7)
```

```python
def gini(row):
    val=classcount(row)
    imp=1
    for i in val:
        prob=(val[i]/len(row))**2
        imp -=prob
        
    return imp
        
```

```python
gini(df1)
```

```python
def partition(row,question):
    tt=[]
    ff=[]
    for i in row:
        if question.match(i):
            tt.append(i)
        else:
            ff.append(i)
    return np.array(tt),np.array(ff)
```

```python
tt,ff=partition(df1,question(3,0.7))
```

```python
ff
```

```python
def info_gain(left, right, current_uncertainty):
    p = float(len(left)) / (len(left) + len(right))
    return current_uncertainty - p * gini(left) - (1 - p) * gini(right)
```

```python
info_gain(tt,ff,0.66666)
```

```python
def best_split(rows):
    bestgain=0
    bestquestion=None
    current_uncertainty=gini(rows)
    n_fe=len(rows[0])-1
    for i in range(n_fe):
        value=set(row[i] for row in rows)
        for val in value:
            ques=question(i,val)
            true_rows, false_rows = partition(rows, ques)
            if len(true_rows) == 0 or len(false_rows) == 0:
                continue
            gain = info_gain(true_rows, false_rows, current_uncertainty)
           
            if gain>=bestgain:
                bestgain,bestquestion=gain,ques
    return bestgain,bestquestion
```

```python
best_split(df1)
```

```python
class decisionnode:
    def __init__(self,question,tb,fb):
        self.question=question
        self.true=tb
        self.false=fb
```

```python
class Leaf:

    def __init__(self, rows):
        self.predictions = classcount(rows)
```

```python
def build_tree(rows):
    gain, question = best_split(rows)

    if gain == 0:
        return Leaf(rows)


    true_rows, false_rows = partition(rows, question)

    true_branch = build_tree(true_rows)

    false_branch = build_tree(false_rows)
    return decisionnode(question, true_branch, false_branch)
```

```python
a=build_tree(df1)
```

```python
def print_tree(node, spacing=""):

    if isinstance(node, Leaf):
        print (spacing + "Predict", node.predictions)
        return

    print (spacing + str(node.question))

    print (spacing + '|--> True:')
    print_tree(node.true, spacing + "  ")

    print (spacing + '|--> False:')
    print_tree(node.false, spacing + "  ")
```

```python
print_tree(a)
```

```python
def classify(row, node):

    if isinstance(node, Leaf):
        return node.predictions
    if node.question.match(row):
        return classify(row, node.true)
    else:
        return classify(row, node.false)
```

```python
def print_leaf(counts):
    total = sum(counts.values()) * 1.0
    probs = {}
    for lbl in counts.keys():
        probs[lbl] = str(int(counts[lbl] / total * 100)) + "%"
    return probs
```
